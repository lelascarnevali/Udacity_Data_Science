# Multiple Inheritance

In this notebook you will define classes for multiple inheritance. 

First, run the cell below to redefine the `DiscountAd` and `StandardAd` classes created in the previous exercise

```python
# Run this cell unchanged
class DiscountAd:
    
    @property
    def discount(self):
    
        raise NotImplementedError
    
    def ad(self):
    
        discount = self.discount
        discount = int(discount * 100)
        return f'{discount}%'
        
    def deliver(self):
    
        return self.ad()

class StandardAd(DiscountAd):
    
    discount = .2
    
    def ad(self):
    
        percent_off = super().ad()
        
        return f'Buy now for {percent_off} off!'
```

In the cell below, follow the comments to define two classes
1. `ExperimentMixin`
1. `ExperimentDiscount40`

```python
import numpy as np
    
# Define a class called `ExperimentMixin`
#### YOUR CODE HERE
class ExperimentMixin:
    
    # Create an abstract class attribute 
    # named `treatment` with a float annotation
    #### YOUR CODE HERE
    treatment: float
    

    # Create a class attribute named
    # `probability` set to the list [.5, .5]
    #### YOUR CODE HERE
    probability = [.5, .5]
    
    # Create a class attribute named
    # `options` set to a list
    # with the following string values
    # "treatment", "control"
    #### YOUR CODE HERE
    options = ["treatment", "control"]
    
    # Define an instance method named `discount`
    # with a single parameter.
    # This method should behave like an attribute
    #### YOUR CODE HERE
    @property
    def discount(self):
    
        # No changes needed to the line below
        choice = np.random.choice(self.options, size=1, p=self.probability)
        
        # No changes needed to the line below
        if choice == 'treatment':
            # Return the `treatment` attribute
            return self.treatment
        
        # No changes needed to the line below
        else:
            # Return the `discount` attribute
            # from the parent class
            return super().discount
        

# Define a class called `ExperimentDiscount40`
# This class should use Mixin Inheritence 
# and is a child of the ExperimentMixin and StandardAd classes      
#### YOUR CODE HERE
class ExperimentDiscount40(ExperimentMixin, StandardAd):
    
    # Create a `treatment` class attribute
    # Set the value to the float .4
    #### YOUR CODE HERE
    treatment = .4
    
```
