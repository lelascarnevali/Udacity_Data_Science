```python
# imports
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## 1. Data Acquision

```python
# load image
input_img = cv2.imread ('images/san_francisco_bay.jpg') 

# convert image to RGB and grayscale (matplotlib uses RGB, OpenCV uses BGR)
img_RGB = cv2.cvtColor(input_img, cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(input_img, cv2.COLOR_BGR2GRAY)
```

```python
# display image
plt.imshow(img_RGB)
plt.title(' San Francisco Bay')
```

**TODO**: Split the image in the three color channels and plot them separately in matplotlib. You can use OpenCV's split function - check the documentation if needed. Can you see from the images which color channel is which?

```python
# TODO: YOUR CODE GOES HERE

R, G, B = cv2.split(img_RGB)
plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
plt.imshow(R, cmap='Reds')
plt.title('Red channel')
plt.subplot(1,3,2)
plt.imshow(G, cmap='Greens')
plt.title('Green channel')
plt.subplot(1,3,3)
plt.imshow(B, cmap='Blues')
plt.title('Blue channel') 
```

```python
# we can make a cutout to work with a smaller part of the image (slicing)
cutout_RGB = img_RGB[0:600, 0:1600]
# cutout_RGB = img_RGB[0:img_RGB.shape[0], 0:img_RGB.shape[1]]
cutout_gray = cv2.cvtColor(cutout_RGB, cv2.COLOR_RGB2GRAY)
plt.imshow(cutout_gray, cmap='gray')
```

**TODO**: Flip the image horizontally and vertically and plot the results! Check out OpenCV's flip function for this task.

```python
# TODO: YOUR CODE GOES HERE

flipped_horizontal = cv2.flip(img_RGB, 1)
flipped_vertical = cv2.flip(img_RGB, 0)
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.imshow(flipped_horizontal)
plt.title('Flipped horizontally')
plt.subplot(1,2,2)
plt.imshow(flipped_vertical)
plt.title('Flipped vertically')
```

**TODO**: Draw a red rectangle around the biggest island (Angel Island) in the original image. Check the OpenCV documentation for a suitable function to use!

```python
# TODO: YOUR CODE GOES HERE

# draw a red rectangle around the biggest island (angel island) in the original image
img_with_rectangle = img_RGB.copy()
top_left = (1000, 1000)
bottom_right = (1500, 1500)
color = (255, 0, 0) # red in RGB
thickness = 10
cv2.rectangle(img_with_rectangle, top_left, bottom_right, color, thickness)
plt.imshow(img_with_rectangle)  
```

**TODO**: Resize the result to 400x400 pixels. Check the OpenCV documentation for a suitable function to use!

```python
# TODO: YOUR CODE GOES HERE

resized_img = cv2.resize(img_with_rectangle, (400, 400))
plt.imshow(resized_img)
```
